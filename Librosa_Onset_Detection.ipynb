```python
import rhythm_regression.notebook_tools as nbt

import librosa
# import librosa.display
# import IPython.display as ipd

import warnings
warnings.filterwarnings('ignore')
```

```python
sample_name = 'Sample 95.m4a'
sample_path = f'./data/audio/{sample_name}'

# ipd.Audio(filename=sample_path)
```

```python
audio, sampling_rate = librosa.load(sample_path)

print(sampling_rate)

print(audio.shape)
```

```python
tr = (34,40)

axs = nbt.plot_signal(audio, title=sample_name, time_range=tr)

onsets = librosa.onset.onset_detect(audio, units='time')

for onset in onsets:
    if tr[0] <= onset <= tr[1]:
        axs.axvline(onset, color='k')
```

```python
tr = (34,40)

axs = nbt.plot_amplitude_envelope(audio, title=sample_name, time_range=tr)

onsets = librosa.onset.onset_detect(audio, units='time')

for onset in onsets:
    if tr[0] <= onset <= tr[1]:
        axs.axvline(onset, color='k')
```
